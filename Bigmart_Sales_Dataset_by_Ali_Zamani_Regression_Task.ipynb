```python
# check for missing values
data.isnull().sum()
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
data_new = imputer.fit_transform(data)
dataset = pd.DataFrame(data_new, columns=data.columns)
```

```python
# check for missing values
dataset.isnull().sum()
```

```python
dataset['Item_Fat_Content'].value_counts()
```

```python
dataset.replace({'Item_Fat_Content': {'low fat': 'Low Fat', 'LF': 'Low Fat', 'reg': 'Regular'}}, inplace=True)
dataset['Item_Fat_Content'].value_counts()
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, 1:-1]
y = dataset.iloc[:,-1]
```

```python
X_new = pd.get_dummies(X, columns=['Item_Fat_Content', 'Item_Type','Outlet_Identifier','Outlet_Size','Outlet_Location_Type','Outlet_Type'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared (R2) Score: {r2}")
```
